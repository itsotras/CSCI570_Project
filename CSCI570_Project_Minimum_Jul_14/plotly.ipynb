```python
import matplotlib.pyplot as plt 
import os 

problem_size = []
cpu_basic = []
for file in os.listdir("Outputs"): 
    if file.endswith(".txt"):
        with open(os.path.join("Outputs", file)) as f:
            lines =[]
            for line in f.readlines():
                lines.append(line.strip())
            computed_X = lines[1]
            computed_Y = lines[2]
            time_ms = float(lines[3])
            size = len(computed_X.replace("_","")) + len(computed_Y.replace("_",""))
            problem_size.append(size)
            cpu_basic.append(time_ms)
        

plt.plot(problem_size, cpu_basic, marker='o')
plt.xlabel("Problem Size")
plt.ylabel("CPU Time (ms)")
plt.title("CPU Time vs Problem Size (Basic DP)")
plt.grid(True)
plt.show()
```
