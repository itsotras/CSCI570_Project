```python
import matplotlib.pyplot as plt 
import os 

import re

output_files = ['in1_generated_out.txt', 'in2_generated_out.txt', 'in3_generated_out.txt', 'in4_generated_out.txt', 'in5_generated_out.txt', 'in6_generated_out.txt', 'in7_generated_out.txt', 'in8_generated_out.txt', 'in9_generated_out.txt', 'in10_generated_out.txt', 'in11_generated_out.txt', 'in12_generated_out.txt', 'in13_generated_out.txt', 'in14_generated_out.txt', 'in15_generated_out.txt']

problem_size = []
cpu_basic = []
memory_basic = []
for file in output_files: 
    if file.endswith(".txt"):
        with open(os.path.join("Outputs_Basic", file)) as f:
            lines =[]
            for line in f.readlines():
                lines.append(line.strip())
            computed_X = lines[1]
            computed_Y = lines[2]
            time_ms = float(lines[3].replace("Milliseconds",""))
            memory_size = float(lines[4].replace("Kilobytes",""))
            size = len(computed_X.replace("_","")) + len(computed_Y.replace("_",""))
            problem_size.append(size)
            cpu_basic.append(time_ms)
            memory_basic.append(memory_size)

cpu_eff = []
memory_eff = []
for file in output_files: 
    if file.endswith(".txt"):
        with open(os.path.join("Outputs_Efficient", file)) as f:
            lines =[]
            for line in f.readlines():
                lines.append(line.strip())
            computed_X = lines[1]
            computed_Y = lines[2]
            time_ms = float(lines[3].replace("Milliseconds",""))
            memory_size = float(lines[4].replace("Kilobytes",""))
            cpu_eff.append(time_ms)
            memory_eff.append(memory_size)



print("Problem Size -",problem_size)
print("CPU time (Basic)-",cpu_basic)
print("Memory Size (Basic)-",memory_basic)
print("---")
print("CPU time (Efficient)-",cpu_eff)
print("Memory Size (Efficient)-",memory_eff)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(problem_size, cpu_basic, marker='o', label="Basic CPU")
plt.plot(problem_size, cpu_eff, marker='s', label="Efficient CPU")

plt.xlabel("Problem Size")
plt.ylabel("CPU Time (ms)")
plt.title("CPU Time vs Problem Size (Basic vs Efficient DP)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(problem_size, memory_basic, marker='o', label="Basic Memory")
plt.plot(problem_size, memory_eff, marker='s', label="Efficient Memory")

plt.xlabel("Problem Size")
plt.ylabel("Memory (KB)")
plt.title("Memory Usage vs Problem Size (Basic vs Efficient DP)")
plt.grid(True)
plt.legend()
plt.show()
```
